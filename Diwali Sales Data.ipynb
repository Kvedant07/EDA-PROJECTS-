```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Diwali Sales Data.csv', encoding='latin1')
```

### Cleaning

```python
data.info()
```

```python
data.describe()
```

```python
data.head(10)
```

```python
data.tail(10)
```

```python
data = data.drop_duplicates()
```

```python
data.drop(columns=['Status', 'unnamed1'], inplace=True, errors='ignore')
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.dtypes
```

```python
data['Amount'] = data['Amount'].astype('int')
```

```python
data['Amount'].dtype
```

```python
data.rename(columns= {'Marital_Status':'Shaddi'} )
```

```python
data.describe()
```

### EDA (Exploratory Data Analysis)

```python
data.columns
```

```python
ax = sns.countplot(x='Gender', data=data)
for bars in ax.containers:                          #value on the bar 
    ax.bar_label(bars)
```

```python
data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales = data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales)
```

```python
ax = sns.countplot(x='Age Group', data=data, hue='Gender')
for bars in ax.containers:                          #value on the bar 
    ax.bar_label(bars)
```

```python
sales = data.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales)
```

```python
state_sales = data.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
# Corrected sns.set() syntax
sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(x='State', y='Orders', data=state_sales)
```

```python
state_sales = data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
# Corrected sns.set() syntax
sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(x='State', y='Amount', data=state_sales)
```

```python
ax = sns.countplot(x='Marital_Status', data=data)

for bars in ax.containers:                          #value on the bar 
    ax.bar_label(bars)
```

```python
state_sales = data.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize': (6, 5)})
sns.barplot(x='Marital_Status', y='Amount', hue='Gender', data=state_sales)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Occupation', data=data)

for bars in ax.containers:                          #value on the bar 
    ax.bar_label(bars)
```

```python
state_sales = data.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize': (18, 5)})
sns.barplot(x='Occupation', y='Amount', data=state_sales)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Product_Category', data=data)

for bars in ax.containers:                          #value on the bar 
    ax.bar_label(bars)
```

```python
state_sales = data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize': (18, 5)})
sns.barplot(x='Product_Category', y='Amount', data=state_sales)
```
