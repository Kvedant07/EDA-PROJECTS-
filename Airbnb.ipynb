```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Airbnb_Open_Data.csv')
```

```python
data.info()
```

```python
data = data.drop_duplicates()
```

```python
data.nunique()
```

```python
data = data.dropna(subset=['NAME','neighbourhood','service fee'])
```

```python
data['service fee'] = data['service fee'].replace('[\$,]', '',regex=True).astype(float)
```

```python
data['last review'] = pd.to_datetime(data['last review'], errors='coerce')
```

```python
data['reviews per month'] = data['reviews per month'].fillna(0)
```

```python
data.info()
```

```python
data['neighbourhood group'].value_counts()
```

```python
data.groupby('neighbourhood group')['service fee'].mean()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=data, x='neighbourhood group', order=data['neighbourhood group'].value_counts().index, palette='viridis')
plt.title('Listings by Neighborhood Group')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='long', y='lat', hue='neighbourhood group', alpha=0.6)
plt.title('Airbnb Locations in City')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=data, x='neighbourhood group', y='service fee')
plt.title('Service Fee Distribution by Neighborhood Group')
plt.xticks(rotation=45)
plt.show()
```
